```python
import pandas as pd
import numpy as np
import os

import plotly.express as px
pd.options.plotting.backend = 'plotly'
```

### Cleaning and EDA

```python
recipes_fp = os.path.join('data', 'RAW_recipes.csv')
recipes_df = pd.read_csv(recipes_fp)
interactions_fp = os.path.join('data', 'RAW_interactions.csv')
interactions_df = pd.read_csv(interactions_fp)
interactions_df
```

```python
recipes_df
```

```python
# merge these two datafrmaes on the recipe id column (left merge) recipes, interactions

merged_recipe_df = recipes_df.merge(interactions_df, how='left', left_on = 'id', right_on = 'recipe_id')
merged_recipe_df
```

```python
# fill the zero ratings with np.nan

merged_recipe_df['rating'].replace(0, np.nan, inplace = True)
merged_recipe_df.head(20)
```

```python
# find the average rating for each recipe

avg_rating_by_recipe = merged_recipe_df.groupby(by = ['id']).mean()['rating']
avg_rating_by_recipe
```

```python
avg_rating_by_recipe.to_frame()
```

```python
# merged_recipe_df['id'].nunique()
```

```python
# add the average rating per recipe back to the dataframe

merged_recipe_df = merged_recipe_df.merge(avg_rating_by_recipe.to_frame().rename(columns = {'rating':'average rating of recipe'}), left_on = 'id', right_index=True)
merged_recipe_df
```

```python
# splitting up the nutrition column

merged_recipe_df['nutrition'] = merged_recipe_df['nutrition'].apply(lambda x : x[1:-1])

merged_recipe_df['nutrition'] = merged_recipe_df['nutrition'].str.split(", ")

merged_recipe_df['calories'] = merged_recipe_df['nutrition'].str[0].astype('float')
merged_recipe_df['total fat'] = merged_recipe_df['nutrition'].str[1].astype('float')
merged_recipe_df['sugar'] = merged_recipe_df['nutrition'].str[2].astype('float')
merged_recipe_df['sodium'] = merged_recipe_df['nutrition'].str[3].astype('float')
merged_recipe_df['protein'] = merged_recipe_df['nutrition'].str[4].astype('float')
merged_recipe_df['saturated fat'] = merged_recipe_df['nutrition'].str[5].astype('float')
merged_recipe_df['carbohydrates'] = merged_recipe_df['nutrition'].str[6].astype('float')

merged_recipe_df.drop(columns=['nutrition'], inplace = True)

```

```python
merged_recipe_df['sugar to calorie ratio'] = merged_recipe_df['sugar'] / merged_recipe_df['calories']

merged_recipe_df
```

```python
merged_recipe_df['review chars'] = merged_recipe_df['review'].str.len()
merged_recipe_df['review word count'] = merged_recipe_df['review'].str.split().str.len()
# merged_recipe_df['review']
```

```python
def posneghelper(x):
    if x < 3:
        return 'Negative'
    if x > 3:
        return 'Positive'
    else:
        return 'Neither'


merged_recipe_df['positive or negative'] = merged_recipe_df['rating'].apply(posneghelper)

merged_recipe_df
```

```python


# merged_recipe_df[['positive or negative', 'calories']]

grouped_ser = merged_recipe_df.groupby(by=['positive or negative']).mean()['sugar to calorie ratio']
observed_diff = grouped_ser['Negative'] - grouped_ser['Positive']

def calculate_diff(df):
    grouped_ser = df.groupby(by=['Shuffled']).mean()['sugar to calorie ratio']
    return grouped_ser['Negative'] - grouped_ser['Positive']

simulated_diffs = []

df = merged_recipe_df[['positive or negative', 'sugar to calorie ratio']]
for i in range(100):
    with_shuffled = df.assign(Shuffled=np.random.permutation(df['positive or negative']))
    simulated_diffs.append(calculate_diff(with_shuffled))

# simulated_diffs
print(simulated_diffs)
p_val = (np.array(simulated_diffs) >= observed_diff).mean()

p_val
```
