```python
import pandas as pd
import numpy as np
import os

import plotly.express as px
pd.options.plotting.backend = 'plotly'
```

### Cleaning and EDA

```python
# read in the two files

recipes_fp = os.path.join('data', 'RAW_recipes.csv')
recipes_df = pd.read_csv(recipes_fp)
interactions_fp = os.path.join('data', 'RAW_interactions.csv')
interactions_df = pd.read_csv(interactions_fp)
```

```python
# merge two dataframes

merged_recipe_df = recipes_df.merge(interactions_df, how='left', left_on = 'id', right_on = 'recipe_id')
```

```python
# replace all zero ratings with nan, since ratings can only be 1-5

merged_recipe_df['rating'].replace(0, np.nan, inplace = True)
```

```python
# find the average rating for each recipe

avg_rating_by_recipe = merged_recipe_df.groupby(by = ['id']).mean()['rating']
```

```python
# add average ratings back to the dataframe

merged_recipe_df = merged_recipe_df.merge(avg_rating_by_recipe.to_frame()
.rename(columns = {'rating':'average rating of recipe'}), left_on = 'id', right_index=True)
```

```python
# splitting up, then removing the nutrition column

merged_recipe_df['nutrition'] = merged_recipe_df['nutrition'].apply(lambda x : x[1:-1])
merged_recipe_df['nutrition'] = merged_recipe_df['nutrition'].str.split(", ")

merged_recipe_df['calories'] = merged_recipe_df['nutrition'].str[0].astype('float')
merged_recipe_df['total fat'] = merged_recipe_df['nutrition'].str[1].astype('float')
merged_recipe_df['sugar'] = merged_recipe_df['nutrition'].str[2].astype('float')
merged_recipe_df['sodium'] = merged_recipe_df['nutrition'].str[3].astype('float')
merged_recipe_df['protein'] = merged_recipe_df['nutrition'].str[4].astype('float')
merged_recipe_df['saturated fat'] = merged_recipe_df['nutrition'].str[5].astype('float')
merged_recipe_df['carbohydrates'] = merged_recipe_df['nutrition'].str[6].astype('float')

merged_recipe_df.drop(columns=['nutrition'], inplace = True)
```

```python
# add sugar to calorie ratio column (what we are interested in looking at)

merged_recipe_df['sugar to calorie ratio'] = merged_recipe_df['sugar'] / merged_recipe_df['calories']
```

#### SOME EXPLORATORY DATA ANALYSIS

```python
# add review word count

merged_recipe_df['review word count'] = merged_recipe_df['review'].str.split().str.len()
```

```python
# adding a column which indicates if the review is positive (4-5 stars) or negative (1-2 stars)
# or neither (3 stars, which we will ignore)

def posneghelper(x):
    if x < 3:
        return 'Negative'
    if x > 3:
        return 'Positive'
    else:
        return 'Neither'


merged_recipe_df['positive or negative'] = merged_recipe_df['rating'].apply(posneghelper)
```
